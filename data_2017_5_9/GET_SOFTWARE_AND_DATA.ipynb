```python
from ftplib import FTP
import os
import time

local_imdb_dir = '/home/idies/workspace/persistent/movies/imdb/'
data_dir = 'data-'+time.strftime('%Y-%m-%d')
if not (os.getcwd() == local_imdb_dir):
    os.chdir(local_imdb_dir)
if not (os.path.exists(data_dir)):
    os.makedirs(data_dir)
os.chdir(data_dir)
print('In directory: '+os.getcwd())

ftp = FTP('ftp.fu-berlin.de') 
ftp.login()
ftp.getwelcome()
ftp.cwd('pub/misc/movies/database/')
filenames = []
for filename in ftp.nlst():
    if ((filename == 'README') or (str(filename)[-3:] == '.gz')):
        filenames.append(filename)

print(str(len(filenames))+' files found.')

i = 1
for filename in filenames:
    local_filename = filename#os.path.join(data_dir, filename)
    file = open(local_filename, 'wb')
    print('Getting file '+str(i)+' of '+str(len(filenames))+': '+local_filename+'...')
    ftp.retrbinary('RETR '+ filename, file.write)
    file.close()
    i = i + 1
ftp.quit()
```

```python
import gzip
#import os
#os.chdir(data_dir)

gzfiles = []
for filename in filenames:
    if(str(filename[-3:])=='.gz'):
        gzfiles.append(filename)

for filename in gzfiles:
    newfilename = str(filename[:-3])
    with gzip.open(filename,'rt',encoding=' ISO-8859-1') as f:
        thisfile = f.read()
        thisfile.encode('utf-8')
        print("writing file: "+newfilename)
        newfile = open(newfilename,'w',encoding='utf-8')
        newfile.write(thisfile)
        newfile.close()
        print('File written!')
        print('\n')
for filename in gzfiles:
    os.remove(filename)
```
