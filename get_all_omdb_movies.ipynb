```python
import urllib.request
import json
import pandas
import numpy as np
import os
from pprint import pprint
import time
os.chdir('/home/idies/workspace/Temporary/raddick/movies_scratch/omdb_all')
#print('ok')
os.getcwd()
```

```python
start_time = time.time()

start = 8999
end = 9000

waiter = 120

omdb_data = {}
exceptions = list()
id_list = range(start*1000, (end*1000)+1)

totalmovies = max(id_list)

print('{0:,.0f} total movies'.format(totalmovies))

for i in id_list:
    if (np.mod(i, 1000) == 0):
        print('Processing movie {0:,.0f} of {1:,.0f}...'.format(i,totalmovies))
        #print('Waiting')
        time.sleep(waiter)
        #print('Waited {0:.0f} seconds'.format(waiter))
    id_string = 'tt'    
    needed_zeroes = 7 - len(str(i))
    for j in range(0,needed_zeroes):
        id_string = id_string + '0'
    id_string = id_string + str(i)
    omdburl = 'http://private.omdbapi.com/?i={0}&apikey=949d45c1'.format(id_string)
    try:
        omdbpage = urllib.request.urlopen(omdburl).read().decode('utf-8')
        omdb_data[i] = json.loads(omdbpage,encoding='utf-8')
        omdb_data[i]['retrieved_at'] = time.time()
    except:
        exceptions.append(i)
        pass

omdb_df = pandas.DataFrame.from_dict(omdb_data,orient='index')
omdb_df.index.name = 'imdb_id'

savefile = 'omdb_{0:.0f}k_{1:.0f}k.csv'.format(start, end)
omdb_df.to_csv(savefile, encoding='utf-8')
exceptions_df = pandas.DataFrame(data=exceptions, columns=['imdb_id'])
exceptfile = 'exceptions/omdb_{0:.0f}k_{1:.0f}k_exceptions.csv'.format(start,end)
exceptions_df.to_csv(exceptfile, encoding='utf-8')

end_time = time.time()

print('Completed in {0:,.0f} seconds!'.format(end_time-start_time))

omdb_df.dropna().size#.sample(10)
omdb_data
```
